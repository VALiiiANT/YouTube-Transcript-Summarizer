Importing the required libraries

```python
from transformers import pipeline
from youtube_transcript_api import YouTubeTranscriptApi as yt #using alias
```

```python
#Enter the url for transcription (here this url is taken for the instance)
url="https://www.youtube.com/watch?v=Tuw8hxrFBH8"
id = url.split("=")[1] #index which is to be kept is specified
transcript=yt.get_transcript(id) #getting the transcript
#the tramscript so obrained is a json obj i.e we will get a dictionary of lists
transcript[0:7]
```

```python
# now we will iterate over the above result to extract the text into a usable form
txt=" "
for i in transcript:
  txt+= ' ' +i['text'] #here i[text] indicates to get the text of i'th line from the dictionary
len(txt) #getting the length of our transcript
```

```python
summarizer= pipeline('summarization') #to create the summary
```

```python
itr=int(len(txt)/1000)  #dividing the result text into chunks for sumaarization
summary=[] #an empty list that will contain the summary

for i in range(0,itr+1):  #this will iterate on the chunk of code summarize and combine each of them
  s=0
  s=i*1000
  e=(i+1)*1000
  out=summarizer(txt[s:e])
  out=out[0]
  out=out['summary_text']
  summary.append(out)

print(summary)
#the code takes a little while to run (as it depends on the length of the video and takes its time to summarize it chunk by chunk)
```
